```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the undersampled data from the CSV file
undersampled_data = pd.read_csv('oversamplingsmote.csv')

# Compute the correlation matrix
correlation_matrix = undersampled_data.corr()

# Generate a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
df=undersampled_data
```

```python
df.describe()
```

```python
df.info()
```

```python
# Columns to retain
columns_to_keep = ['V13', 'V8', 'V15', 'V20']

# Identify the columns to drop (v2 to v22, excluding the ones in columns_to_keep)
columns_to_drop = [col for col in df.columns if col.startswith('V') and 2 <= int(col[1:]) <= 22 and col not in columns_to_keep]

# Drop the unwanted columns
df = df.drop(columns=columns_to_drop)

print("Remaining columns:", df.columns)
```

```python
correlation_matrix = df.corr()
# Generate a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# Save the DataFrame with selected columns to a CSV file
df.to_csv('filtered_data.csv', index=False)

print("Data saved to 'filtered_data.csv'")
```
